1. Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the
project where required.


 -  



```
**PROJECT FILE STRUCTURE**

*/css-assignment*

*|---index.html*

*|---style.css*


***HTML FILE (index.html)***


<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>CSS Methods Example</title>
  <!-- Internal CSS -->
    <style>
        .internal {
            color: blue;
            font-size: 20px;
        }
    </style>
 <!-- Link to External CSS -->
    <link rel="stylesheet" href="style.css">
</head>
<body>
 <h2>Different Ways to Apply CSS</h2>
 <!-- Inline CSS -->
    <div class="box" style="color: red; font-weight: bold;">
        This div uses Inline CSS (Red text, Bold)
    </div>
 <!-- Internal CSS -->
    <div class="internal">
        This div uses Internal CSS (Blue text, 20px font size)
    </div>
 <!-- External CSS -->
    <div class="external">
        This div uses External CSS (Green text, Underlined)
    </div>
 <!-- Conflicting CSS (Inline, Internal, External) -->
    <div class="conflict" style="color: purple;">
        Which CSS will apply? (Check style.css for the external definition)
    </div>
</body>
</html>  


***EXTERNAL CSS FILE(style.css)***


<style>
/* External CSS */
.external {
    color: green;
    text-decoration: underline;
}

/* Conflicting CSS */
.conflict {
    color: orange; /* This will be overridden by Inline CSS in index.html */
    font-style: italic;
}
</style>


```

2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.


  -   



```
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.</title>
    
    <style>
        .box {
            border: 2px solid black;
            height: 50vh;
            width: 70vw;
            background-color: rgb(228, 191, 200);
            padding-left: 10px;
        }

        .box_heading {
            color: red;
        }

        .box_para {
            color: blue;
        }

        .box_subheading {
            color: rgb(25, 119, 39);
        }

        .box_subpara {
            color: crimson;
        }

        .box_btn {
            color: blue;
            cursor: pointer;
        }
    </style>
</head>

<body>
    <div class="box">
        <h2 class="box_heading">THIS IS MAIN HEADING</h2>
        <p class="box_para">This is Paragraph</p>
        <h3 class="box_subheading">THIS IS SUB HEADING</h3>
        <p class="box_subpara">This is sub paragraph</p>
        <button class="box_btn">THIS IS A BUTTON</button>
    </div>
</body>

</html>
```

3. Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.

 -  


```

<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title> Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.</title>
    
  <style>
        .first-name {
            background-color: aqua;
            opacity: 0.5;
            /* this is the way to apply opacity. */
        }
        .last-name{
            background-color: blue;
        }
        .email{
            background-color: brown;
        }
        .password{
            background-color: rgb(81, 92, 70);
        }
        .confirm-password{
            background-color: chocolate;
        }
    </style>
</head>
<body>
    <h2>Form</h2>
    <form action="post">
        <div>
            <label for="fname">First Name: </label>
            <input class="first-name" type="text" id="fname" name="username" placeholder="First Name" required
                minlength="3"><br><br>
            <label for="lname">Last Name: </label>
            <input class="last-name" type="text" id="lname" name="username" placeholder="Last Name" required
                minlength="3"><br><br>
        </div>
        <div>
            <label for="email">Email: </label>
            <input class="email" type="email" id="email" name="email" placeholder="Email" required minlength="3">
        </div><br>
        <div>
            <label for="password">Password: </label>
            <input class="password" type="password" id="Password" name="password" placeholder="******" required>
        </div><br>
        <div>
            <label for="confirm password">Confirm Password: </label>
            <input class="confirm-password" type="confirm password" id="confirm Password" name="confirm password" placeholder="******"
                required>
        </div><br>
        <div>
            <label for="age">Age:</label>
            <input type="number" id="age" name="age" required max="150" min="0" value="0" >
        </div><br>
        <div>
            <label for="male">Male:</label>
            <input type="radio" id="male" name="gender" required value="male" checked >
            <label for="female">Female:</label>
            <input type="radio" id="female" name="gender" required value="female">
        </div><br>
        <div><label for="terms">I agree to terms & conditions</label>
            <input type="checkbox" id="terms" name="terms" checked>
        </div><br>
        <button type="submit">Register</button>
    </form>
    
</body>

</html>
```
